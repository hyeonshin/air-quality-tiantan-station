## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana korelasi antar variabel terhadap kualitas udara?

```python
data = tiantan_df

numeric_columns = data.select_dtypes(include='number').drop(columns="No", errors='ignore')

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(10, 8)) 
sns.heatmap(
    correlation_matrix,
    annot=True,          
    cmap='coolwarm',      
    fmt=".2f",            
    linewidths=0.5         
)
plt.title('Heatmap of Correlations')
plt.show()
```

**Insight:**
- Korelasi antara PM2.5 dengan PM10, NO2, dan CO menunjukkan bahwa semakin tingginya PM2.5 sebanding PM10, NO2, dan CO yang menunjukkan bahwa seberapa buruk kualitas udara berdasarkan keterkaitan ketiga variabel tersebut.
- TEMP juga menunjukkan korelasi terhadap DEWP
- RAIN juga menenjukkan korelasi terhadap DEWP

### Pertanyaan 2:

```python
yearly_tiantan_df = tiantan_df.groupby("year").mean(numeric_only=True)

plt.figure(figsize=(10, 5))
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["PM2.5"], label="PM2.5",
    marker='o', 
)
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["PM10"], label="PM10",
    marker='o', 
)
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["SO2"], label="SO2",
    marker='o', 
)
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["NO2"], label="NO2",
    marker='o', 
)
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["O3"], label="O3",
    marker='o', 
)
plt.title("Angka Polutan Udara di Tiantan per Tahun", loc="center", fontsize=16)
plt.xlabel("Tahun")
plt.ylabel("mg/m³")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

```python
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["CO"], label="CO",
    marker='o', 
)
plt.title("Angka Polutan CO di Tiantan per Tahun", loc="center", fontsize=10)
plt.xlabel("Tahun")
plt.ylabel("mg/m³")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

**Insight:**
- Pada tiga tahun pertama data 2013 hingga 2015 data polutan cenderung stabil.
- Lalu pada tahun 2016 terdapat penuruan jumlah polutan.
- Namun pada tahun berikutnya yaitu 2017 mengalami lonjakan kenaikan jumlah polutan.

### Pertanyaan 3: Pada tahun berapa kualitas udara terbaik dan terburuk?

```python
plt.plot(
    yearly_tiantan_df.index, yearly_tiantan_df["PM2.5"], label="PM2.5",
    marker='o', 
)
plt.title("Angka Polutan PM2.5 di Tiantan per Tahun", loc="center", fontsize=10)
plt.xlabel("Tahun")
plt.ylabel("mg/m³")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend()
plt.show()
```

**Insight:**
- Pada tahun 2016 menjadi tahun terbaik dalam hal kualitas udara karena memiliki jumlah polutan yang rendah.
- Pada tahun 2017 menjadi tahun terburuk dalam hal kualitas udara karena memiliki jumlah polutan yang tinggi.

### Pertanyaan 4: Pada jam berapakah kualitas udara menjadi sangat buruk?

```python
hourly_pm25 = data.groupby('hour')['PM2.5'].mean()

max_hour = hourly_pm25.idxmax()
max_value = hourly_pm25.max()

colors = ['red' if hour == max_hour else 'gray' for hour in hourly_pm25.index]

plt.figure(figsize=(10, 6))
hourly_pm25.plot(kind='bar', color=colors, edgecolor='black')
plt.title('Average PM2.5 Levels by Hour', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=14)
plt.ylabel('PM2.5 Concentration', fontsize=14)
plt.xticks(rotation=0)  
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.text(max_hour, max_value + 2, f'{max_value:.2f}', ha='center', va='bottom', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

**Insight:**
- Tingkat polutan tertinggi berdasarkan rata-rata PM2.5 perjamnya berada pada jam 00 yang mencapai angka 88,26.

## Analisis Lanjutan

### Clustering PM2.5

```python
def assign_cluster(pm25):
    if pm25 <= 15:
        return 'Good'
    elif pm25 <= 65:
        return 'Moderate'
    elif pm25 <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif pm25 <= 250:
        return 'Unhealthy'
    elif pm25 <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

tiantan_df['Cluster'] = tiantan_df['PM2.5'].apply(assign_cluster)
tiantan_df[["PM2.5","Cluster"]].sample(25)
```

**Insight:**
- Setelah dilakukan teknik Cluster berdasarkan berdasarkan beberapa kategori dengan aturan tertentu, didapatkan hasil sebagai berikut:

```python
tiantan_df.groupby('Cluster')['PM2.5'].count()
```

```python
import matplotlib.pyplot as plt

cluster_counts = tiantan_df.groupby('Cluster')['PM2.5'].count()

max_value = cluster_counts.max()
max_index = cluster_counts.idxmax()

colors = ['red' if value == max_value else 'gray' for value in cluster_counts.values]

plt.barh(cluster_counts.index, cluster_counts.values, color=colors, edgecolor='black')

plt.xlabel('PM2.5 Count')
plt.ylabel('Cluster')
plt.title('PM2.5 Count by Cluster')

plt.text(max_value, max_index, f'{max_value}', 
         va='center', ha='left', fontsize=10, color='black')

plt.tight_layout()

plt.show()
```

**Insight:**
- Dari hasil clustering PM2.5 didapatkan bahwa jumlah PM2.5 tingkat sedang paling banyak terjadi,
- Kemudian disusul dengan tingkat yang tidak sehat untuk grup yang sensitif.
